### 1.2 Which word has the greatest number of distinct tags? What are the tags, and which parts of speech do they represent?

```python
#1.2

words_list = []
tags_list = []
for sentence in train_data:
    for token in sentence:
        word = token.form
        tag = token.upos
        words_list.append(word)
        tags_list.append(tag)
        
dictionary = dict(zip(words_list, tags_list)) #creating a dictionary of the word and tag
words_nd_tags_lists = [(k, v) for k, v in dictionary.items()]
print(words_nd_tags_lists)
```

```python
#1.2
import nltk
data = nltk.ConditionalFreqDist((word.lower(), tag) for (word, tag) in words_nd_tags_lists)
for word in sorted(data.conditions()):
    if len(data[word]) > 3:
        tags = [tag for (tag, _) in data[word].most_common()]
        print("Word with greater than 3 tags: " ,word, ' '.join(tags))
    
```

```python
#1.2
for word in sorted(data.conditions()):
    if len(data[word]) > 2:
        tags = [tag for (tag, _) in data[word].most_common()]
        print("Word with greater than 2 tags: " , word, ' '.join(tags))
```

```python
#1.2
for word in sorted(data.conditions()):
    if len(data[word]) > 1:
        tags = [tag for (tag, _) in data[word].most_common()]
        print("Word with greater than 1 tag: " , word, ' '.join(tags))
```

Looking at the results above, there are no words with greater than 3 tags. There are two words "as" and "press" with greater than 2 tags, and there are many words with greater than 1 tag. Thus, the answer to 1.2 is the words "as" and "press" have the greatest number of distinct tags.

#### 1.4 Which tags are nouns most commonly found after, and which parts of speech do they represent?

```python
#1.4
tags_bigram =nltk.bigrams(tag for (word,tag)  in  words_nd_tags_lists)

after_noun = nltk.FreqDist(tag1 for (tag1,tag2)  in  tags_bigram if tag2 == 'NOUN')

print (after_noun.most_common(20))
```

NOUN, VERB, ADJ, PROPN, ADV, and NUM, among others shown above, are the parts of speech tags for which nouns are most commonly found after.

### 2. Generate some statistics for tagged data to answer the following questions:

#### 2.1 What proportion of word types are always assigned the same part-of-speech tag?

```python
#2.1
data = nltk.ConditionalFreqDist((word.lower(), tag) for (word, tag) in words_nd_tags_lists)
single_tags_count = 0
nonsingle_tags_count = 0
for word in sorted(data.conditions()):
    if len(data[word]) == 1:
        single_tags = [tag for (tag, _) in data[word].most_common()]
        single_tags_count = single_tags_count + 1
    else:
        nonsingle_tags_count = nonsingle_tags_count + 1

proportion_single_tags = single_tags_count / (single_tags_count+nonsingle_tags_count)  
print("Proportion of word types that're assigned the same part-of-speech tag: ", proportion_single_tags)
```

#### 2.2 How many words are ambiguous, in the sense that they appear with at least two tags?

```python
#2.2
at_least_two_tags_ct = 0
other_words_count = 0
for word in sorted(data.conditions()):
    if len(data[word]) >= 2:
        at_least_two_tags = [tag for (tag, _) in data[word].most_common()]
        #print("Word with greater than 1 tag: " , word, ' '.join(at_least_two_tags))
        at_least_two_tags_ct = at_least_two_tags_ct + 1
    else:
        other_words_count = other_words_count + 1
print("No. of words that are ambiguous (at least with 2 tags): ",at_least_two_tags_ct)
```

#### 2.3 What percentage of word tokens in the UD Corpus involve these ambiguous words?

```python
percent_ambiguous_words = at_least_two_tags_ct / (at_least_two_tags_ct+other_words_count)
print(percent_ambiguous_words, "percentage of word tokens in the UD Corpus involve these ambiguous words.")
```

### 3. Since you are used to the corpus, you will build a Most Frequent tagger. The most frequency tagger does not take context into account. It is a simple dictionary based tagger that assigns the most frequent tag to a word. You will get the word-tag counts from the downloaded corpora. #### 3.1 You will write a function that takes a sentence and the dictionary as input. The output should consist of a tag for each word in the sentence. Repeat the step for all the sentences in the test corpus.

#### The Most Frequent Tagger

```python
#3.1
directory_train = "/Users/shristidhakal/Documents/Grad School/Comp Linguistics/week 10/UD_English-GUM-master/en_gum-ud-train.conllu"

train_data = pyconll.load_from_file(directory_train)
tags_list = []
tagged_words_list = []
for sentence in train_data:
    sentence_text = sentence.text
    tokenized_words = nltk.word_tokenize(sentence_text)
    tagged_words = nltk.pos_tag(tokenized_words)
    tagged_words_list.append(tagged_words)
    tags = [tag for (word, tag) in tagged_words]
    tags_list.append(tags)
most_freq_tag = nltk.FreqDist(tags).max()
most_freq_tag
```

```python
#3.1
def most_frequent_tagger(directory_train):
    words_list = []
    tags_list = []
    for sentence in train_data:
        for token in sentence:
            word = token.form
            tag = token.upos
            words_list.append(word)
            tags_list.append(tag)
    
    dictionary = dict(zip(words_list, tags_list))    
    words_nd_tags_lists = [(k, v) for k, v in dictionary.items()]
    
    new_words_list = []
    new_tags_list = []
    for (word, tag) in words_nd_tags_lists:
        tagger = "NNP"
        word = word
        new_words_list.append(word)
        new_tags_list.append(tagger)
    
    dictionary = dict(zip(new_words_list, new_tags_list))    
    most_frequent_tag = [(k, v) for k, v in dictionary.items()]
    
    return most_frequent_tag

most_frequent_tagger(directory_train)
```
