## In Pirates of the Carribean, how could you find the things said by a particular character (e.g., JACK SPARROW)?

```python
import nltk
from nltk.corpus import webtext
webtext.fileids()
```

```python
from notebook.services.config import ConfigManager
cm = ConfigManager().update('notebook', {'limit_output': 5})

pirates = webtext.raw(fileids='pirates.txt')
print(pirates[0:3999]) #printing the first 4000 characters
```

```python
import re
for line in re.findall('JACK SPARROW.+', pirates):
    print(line)
```
